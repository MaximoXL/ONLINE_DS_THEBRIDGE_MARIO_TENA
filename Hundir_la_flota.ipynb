```python
import numpy as np
import random
#Cramos el tablero
celda_vacia = "-"
celda_barco = "O"
celda_disparo = "X"
celda_agua = "~"
list_orientacion = ["N" , "S", "E", "O"]

def crear_tablero():
    tablero = np.full((10, 10), celda_vacia)
    return tablero
tablero = crear_tablero()
print(tablero)


def coloca_barco_plus(tablero, barco):
    """
    Esta función coloca un barco en el tablero si cabe.
    Si no cabe, retorna False.
    """
    for fila, columna in barco:
        if fila < 0 or fila >= tablero.shape[0] or columna < 0 or columna >= tablero.shape[1]:
            return False  # Si el barco se sale del tablero
        if tablero[fila, columna] != celda_vacia:
            return False  # Si la celda ya está ocupada
    # Colocar el barco
    for fila, columna in barco:
        tablero[fila, columna] = celda_barco
    return tablero

def crea_barco_aleatorio(tablero, eslora=4, num_intentos=100):
    """
    Función que crea un barco aleatorio en el tablero.
    Si no puede colocarse después de num_intentos, retorna None.
    """
    num_max_filas = tablero.shape[0]
    num_max_columnas = tablero.shape[1]
    
    # Intentar colocar el barco durante 'num_intentos' intentos
    for intento in range(num_intentos):
        barco = []
        pieza_original = (random.randint(0, num_max_filas - 1), random.randint(0, num_max_columnas - 1))
        print(f"Intento {intento+1}: Pieza original: {pieza_original}")
        barco.append(pieza_original)
        orientacion = random.choice(["N", "S", "O", "E"])
        print(f"Con orientación: {orientacion}")
        
        fila = pieza_original[0]
        columna = pieza_original[1]
        
        # Generamos las celdas del barco según la orientación
        for i in range(eslora - 1):
            if orientacion == "N":
                fila -= 1
            elif orientacion == "S":
                fila += 1
            elif orientacion == "E":
                columna += 1
            else:  # orientacion == "O"
                columna -= 1
            
            pieza = (fila, columna)
            barco.append(pieza)
        
        # Verificar si el barco cabe en el tablero
        tablero_temp = coloca_barco_plus(tablero, barco)
        if isinstance(tablero_temp, np.ndarray):  # Si el barco se coloca correctamente
            return tablero_temp
        
        print(f"No pude colocar el barco en el intento {intento+1}, probando otra vez...")
    
    # Si no se pudo colocar después de 'num_intentos', devolver None
    print(f"No pude colocar el barco después de {num_intentos} intentos.")
    return None

# Crear un tablero vacío de 10x10
tablero = np.full((10, 10), celda_vacia)

# Intentar colocar un barco aleatorio de tamaño 4 en el tablero
tablero = crea_barco_aleatorio(tablero, eslora=4, num_intentos=100)

# Mostrar el tablero resultante
print(tablero)
```
